```python
a= int(input("enter a:"))
b= int(input("enter b:"))

while True:
  try:
    c= int(input("enter c:"))
    break # Exit loop if input is a valid integer
  except ValueError:
    print("Invalid input. Please enter an integer.")

if a>b and a>c:
  print("a is greatest")
elif b>a and b>c:
  print("b is greatest")
else:
  print("c is greatest")
```

```python
data={
    "emp_id":[1,2,3,4,5],
    "emp_name":["amit","sumit","puja","rani","hena"],
    "emp_salary":[20000,50000,40000,30000,90000],
    "emp_feedback":["good","very good","good","excellent","avg"]
}
df1=pd.DataFrame(data)
df1
```

```python
df1["emp_name"].value_counts()
```

```python
df1["emp_feedback"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder , LabelEncoder
```

```python
One=OneHotEncoder(sparse_output=False)
L=LabelEncoder()
```

```python
df1['emp_feedback']=L.fit_transform(df1['emp_feedback'])
```

```python
df1
```

```python
cat_col=df1.select_dtypes(include=["object"]).columns.tolist()
```

```python
a=One.fit_transform(df1[cat_col])
```

```python
df1_enc=pd.DataFrame(a,columns=One.get_feature_names_out(cat_col))
df1_enc
```
